###Mutable And Immutable Data Types :

* Mutable : These variables values can be modified.
  
  * Examples : Lists, dictionary and Sets

* Immutable : These values cannot be modified once it was declared or created.

  * Examples : Intergers, Float, Complex Numbers, Strings, Tuples.

###Mutable List:

```python
list_mutable = [1,2,3]
```

```python
print("Before Changes :",list_mutable)
print("Before Changes, Id :",id(list_mutable))
```

```python
list_mutable[0] = "nikki"
```

```python
list_mutable.append("2002")
```

* After all operations List will be like :

```python
print("After Changes :",list_mutable)
print("After Changes, Id :",id(list_mutable))
```

###Mutable Dictionary:

```python
dict_mutable = {'name':'Pavan','frnd':'ajay','year':2003}
```

```python
print("Before Changes :",dict_mutable)
print("Before Changes, Id :",id(dict_mutable))
```

```python
dict_mutable['frnd'] = 'nikki'
```

```python
dict_mutable['location'] = 'India'
```

```python
print("After Changes :",dict_mutable)
print("AFter Changes, Id :",id(dict_mutable))
```

###Mutable Sets:

```python
set_mutable = {"pavan",0,"nikki",2,"India"}
```

```python
print("Before Changes :",set_mutable)
print("Before Changes, Id :",id(set_mutable))
```

* Order of elements in a set in Python is not guaranteed. This is because sets in Python are implemented as hash tables, which do not maintain the order of elements. So when we print the order may change sometimes

```python
set_mutable.add("REVA University")
```

```python
set_mutable.remove("India")
```

```python
print("After Changes :",set_mutable)
print("After Id is :",id(set_mutable))
```

###* In all above examples Before and After ID's are always same i.e previous value is being replaced with new values
